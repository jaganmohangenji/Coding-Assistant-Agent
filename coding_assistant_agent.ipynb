```python
from groq import Groq
import os

client = Groq(api_key=os.environ["GROQ_API_KEY"])

def coding_agent():
    print("💻 AI Coding Assistant (type 'exit' to stop)\n")

    while True:
        print("\nChoose option:")
        print("1. Ask coding question")
        print("2. Generate code")
        print("3. Explain code")
        print("4. Debug error")

        choice = input("Enter choice: ")

        if choice == "exit":
            print("Goodbye 👋")
            break

        try:
            if choice == "1":
                question = input("Enter your coding question: ")
                prompt = f"Answer clearly with example:\n{question}"

            elif choice == "2":
                task = input("What code do you want? ")
                prompt = f"Write clean and working code for:\n{task}"

            elif choice == "3":
                code = input("Paste code to explain:\n")
                prompt = f"Explain this code step by step:\n{code}"

            elif choice == "4":
                error = input("Paste error or code:\n")
                prompt = f"Find and fix error in this:\n{error}"

            else:
                print("Invalid choice ❌")
                continue

            response = client.chat.completions.create(
                model="llama-3.1-8b-instant",
                messages=[{"role": "user", "content": prompt}]
            )

            answer = response.choices[0].message.content

            print("\n✅ Result:\n")
            print(answer)
            print("\n" + "="*60)

        except Exception as e:
            print("❌ Error:", e)

# Run agent
coding_agent()
```
